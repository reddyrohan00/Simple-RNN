```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sent=[  'the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good',]
```

```python
#vocab size
vocab_size=10000
```

```python
onehotrep=[one_hot(words,vocab_size)for words in sent]
```

```python
onehotrep
```

```python
#word_embedding
from tensorflow.keras.layers import Embedding
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential
import numpy as np 
import pandas as pd
```

```python
# we pad the sequences
sent_len=8
sents=pad_sequences(onehotrep,padding="pre",maxlen=sent_len)# can use post as well for 0's at the end
print(sents)
```

```python
#we use word2vec
dim=10 #no of feature
model=Sequential()
model.add(Embedding(vocab_size,dim,input_length=sent_len))
model.compile('adam','mse')
```

```python
model.predict(sents)
```
