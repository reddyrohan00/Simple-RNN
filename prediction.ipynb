```python
import numpy as np 
import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing import sequence
from tensorflow.keras.models import load_model
```

```python
model=load_model('simple_rnn.h5')
model.summary()
```

```python
#function to decode reviews
words_index=imdb.get_word_index()
reversed={value:key for key,value in words_index.items()}
def decoded_review(encoded_review):
    return ' '.join([reversed.get(i-3,'?')for i in encoded_review])
```

```python
#function to preprocess user i/p
def preprocess_text(text):
    if isinstance(text, np.ndarray):
        text = text.item()
    words=text.lower().split()
    encoded_review=[words_index.get(word,2)+3 for word in words]
    padded_review=sequence.pad_sequences([encoded_review],maxlen=500)
    return padded_review
```

```python
### prediction function
def predict_sentiment(review):
    preprocessed_input=preprocess_text(review)
    prediction=model.predict(preprocessed_input)
    sentiment='Positive' if prediction[0][0]>=0.5 else 'Negative'
    return sentiment,prediction[0][0]
```

```python
ex="I hated the plot , the characters and all were a bit okay but the twist at the end is disastrous"
sentiment,score=predict_sentiment(ex)
print(sentiment,score)
```
